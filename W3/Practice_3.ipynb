```python
import networkx as nx
import numpy as np
from random import sample
import matplotlib.pyplot as plt
```

## Initialize graph with example nodes and edges.

```python
G = nx.Graph()

G.add_node("Paris")
G.add_node("London")
G.add_node("New York")
G.add_node("Beijing")
G.add_node("Saigon")
G.add_node("Tokyo")
G.add_node("Seoul")
G.add_node("Singapore")
G.add_node("Dubai")

G.add_weighted_edges_from(
    [
        ("Paris", "Tokyo", 9712),
        ("Paris", "New York", 5834),
        ("Beijing", "Singapore", 4470),
        ("Singapore", "Dubai", 5833),
        ("Dubai", "Beijing", 5851),
        ("Saigon", "New York", 14291),
        ("Tokyo", "London", 9559),
        ("Tokyo", "Saigon", 4329),
        ("Saigon", "Seoul", 3608),
        ("Paris", "London", 487),
        ("Seoul", "New York", 11052),
        ("Seoul", "Tokyo", 1159),
        ("Singapore", "New York", 15330),
    ]
)
```

## Drawing the Graph with networkx.

```python
nx.draw(G)
```

## Draw graph with node labels

```python
nx.draw(G, with_labels = True)
```

## Using layout to draw graph
### Spring layout

```python
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels = True)
```

### Circular layout

```python
# set circular layout
pos = nx.circular_layout(G)

# draw graph
nx.draw(G, pos, with_labels = True)
```

### Draw edges with labels

```python
# draw edge labels
edge_labels = {
        ("Seoul","New York"): '3 flights daily', 
        ("Paris","London"): '5 flights daily',
        ("New York","Paris"): '2 flights daily',
    }
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels = True)
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels,
    font_color='green'
)
```

```python
colors = ['green', 'yellow', 'magenta', 'lightblue', 'lightgreen', 'pink', 'orange', 'red', 'blue']
options = {
    'node_color': colors,       # color of node
    'node_size': 3000,          # size of node
    'width': 1,                 # line width of edges
    'arrowsize': 18,            # size of arrow
    'edge_color':'red',         # edge color
}

pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels = True, arrows=True, **options)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
```

## Tree layout

```python
from networkx.drawing.nx_agraph import graphviz_layout

# Define example tree
tree_graph = nx.Graph()
edges = [
        ("A", "B"),
        ("A", "C"),
        ("A", "D"),
        ("B", "E"),
        ("B", "F"),
        ("C", "G"),
        ("D", "H"),
        ("D", "I"),
        ("H", "K"),
        ("H", "L"),
        ("I", "M"),
        ("I", "N"),
    ]
tree_graph.add_edges_from(edges)

# Tree layout
pos=graphviz_layout(tree_graph, prog='dot')
nx.draw(tree_graph, pos, with_labels=True, arrows=False)
```

## Circular tree layout

```python
# Circular tree layout
pos=graphviz_layout(tree_graph, prog='twopi')

nx.draw(tree_graph, pos, with_labels=True, arrows=False)
```

## Automatically generate a balanced tree in networkx

```python
# Define the number of levels and the number of children nodes at each level.
num_levels = 5
num_child_nodes = 2

# Generate a binary balanced tree with 5 layer
G = nx.balanced_tree(num_child_nodes, num_levels)

# Use tree layout from graphviz
pos = graphviz_layout(G, prog="twopi", args="")
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_size=300, alpha=0.5, node_color="green", with_labels=True)
plt.axis("equal")
plt.show()
```

## Draw graph with betweenness centrality
This is a large scale graph. Therefore, it takes a few minutes to run drawing.

```python
# largest connected component
def get_largest_component(G):
    components = nx.connected_components(G)
    largest_component = max(components, key=len)
    largest_subgraph = G.subgraph(largest_component)
    
    return largest_subgraph

def draw_graph_betweenness_centrality(G, largest_component):
    # Calculate betweenness centrality
    centrality = nx.betweenness_centrality(largest_component, k=10, endpoints=True)

    # Compute community structure
    lpc = nx.community.label_propagation_communities(largest_component)
    community_index = {n: i for i, com in enumerate(lpc) for n in com}

    # Draw graph using matplotlib

    fig, ax = plt.subplots(figsize=(20, 15))
    pos = nx.spring_layout(largest_component, k=0.15, seed=4572321)
    node_color = [community_index[n] for n in largest_component]
    node_size = [v * 20000 for v in centrality.values()]
    nx.draw_networkx(
        largest_component,
        pos=pos,
        with_labels=False,
        node_color=node_color,
        node_size=node_size,
        edge_color="gainsboro",
        alpha=0.4,
    )

    # Title/legend
    font = {"color": "blue", "fontweight": "bold", "fontsize": 20}
    ax.set_title("Gene functional association network (C. elegans)", font)
    # Change font color for legend
    font["color"] = "green"

    ax.text(
        0.80,
        0.10,
        "Node color: community structure",
        horizontalalignment="center",
        transform=ax.transAxes,
        fontdict=font,
    )
    ax.text(
        0.80,
        0.06,
        "Node size: betweenness centrality",
        horizontalalignment="center",
        transform=ax.transAxes,
        fontdict=font,
    )

    # Resize figure for label readability
    ax.margins(0.1, 0.05)
    fig.tight_layout()
    plt.axis("off")
    plt.show()
    
# Read graph data from edgelist
G = nx.read_edgelist("data/WormNet.v3.benchmark.txt")
largest_component = get_largest_component(G)
draw_graph_betweenness_centrality(G, largest_component)
```

## Remove  randomly selected nodes and low-degree nodes to run code faster for a large-scale graph

```python
# Remove randomly selected nodes (to make example fast)
def remove_random_selected_nodes(G):
    num_to_remove = int(len(G) / 1.5)
    nodes = sample(list(G.nodes), num_to_remove)
    G.remove_nodes_from(nodes)
    return G

# Remove low-degree nodes
def remove_low_degree_nodes(G):
    low_degree = [n for n, d in G.degree() if d < 10]
    G.remove_nodes_from(low_degree)
    return G

G = remove_random_selected_nodes(G)
G = remove_low_degree_nodes(G)
largest_component = get_largest_component(G)
draw_graph_betweenness_centrality(G, largest_component)
```
