# Ex 1: 
Given an undirected graph with edges = [(1, 2), (1, 6), (2, 3), (2, 4), (2, 6), 
         (3, 4), (3, 5), (4, 8), (4, 9), (6, 7), (8, 9)]
  
1. Draw the given graph
2. Print the total number of nodes in the graph
3. Print the total number of edges in the graph
4. Provide a list of all edges in the graph
5. Find degree of a node 1
6. Determine if the given graph is connected or if there are disconnected components. If disconnected, identify and describe each component
7. Calculate and display the degree distribution for the entire graph. Explain what insights this distribution provides about the structure of the graph

```python
import networkx as nx
import networkx
import numpy as np
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([(1, 2), (1, 6), (2, 3), (2, 4), (2, 6), (3, 4), (3, 5), (4, 8), (4, 9), (6, 7), (8, 9)])

nx.draw_networkx(G, with_labels=True, font_weight='bold', node_color='skyblue', font_color='black', node_size=800)
#2 Total number of nodes
print("Total number of nodes: ", int(G.number_of_nodes()))
#3 Total number of edges
print("Total number of edges: ", int(G.number_of_edges()))
#4 List of all nodes
print("List of all nodes: ", list(G.nodes()))
#5 Find degree of a node 1
degree_of_node_1 = G.degree[1]
print("Degree of node 1:", degree_of_node_1)

#6 Determine if the given graph is connected...
# Check if the graph is connected
if nx.is_connected(G):
    print("The graph is connected")
else:
    print("The graph is not connected")

# Get the connected components
components = nx.connected_components(G)

# Print each component
for i, component in enumerate(components):
    print(f"Component {i+1}: {component}")
#7 Calculate and display the degree distribution for the entire graph
degree_distribution = dict(G.degree())
print("Degree distribution:", degree_distribution)

```

# Ex 2: Using NetworkX, read file practice.txt 

In the practice1.txt, the format of each row: node - node - edge type

Question:
1. Draw the graph

2. Print:
    - The total number of nodes and edges
    - Provide a list of all nodes

3. Print all the neighbours of node David

4. Print all the neighbours of node David who are its friends

5. What is the degree of node David in the graph?

6. What is the relation type between David and Amy?

7. Is the graph a connected graph?

8. What is the average degree of the nodes in the graph?

9. How many nodes in the graph have a degree greater than 3?


```python
import networkx as nx
import networkx
import numpy as np
import matplotlib.pyplot as plt


Graphtype =nx.MultiGraph()
G = nx.read_edgelist('./data/practice1.txt', create_using=Graphtype, data =[('relation', str)])


nx.draw_networkx(G, with_labels=True)

```

```python
#2. Print
#Total number of nodes and edges
print("Total number of nodes: ", int(G.number_of_nodes()))
print("Total number of edges: ", int(G.number_of_edges()))

#List of nodes and edges
print("List of all nodes: ", list(G.nodes()))
print("List of all edges: ", list(G.edges(data = True)))

# 3. Print all the neighbours of node David
print("List of all nodes we can go to in a single step from node David: ", list(G.neighbors('David')))

# 4. Print all the neighbours of node David who are its friends
for i in list(G.neighbors('David')):
    if(G[i]['David'][0]["relation"]=='friend'):
        print(i)

# 5. Degree of node David
print("Degree of node David:",G.degree['David'])

# 6. Relation type between David and Amy
print(G['David']['Amy'][0]['relation'])

# 7. Check if the graph is connected
is_connected = nx.is_connected(G)
print(is_connected)

# 8. Average degree
average_degree = sum(d for n, d in G.degree()) / float(len(G))
print(average_degree)

# 9. Nodes with degree greater than 3
nodes_degree_gt_3 = [node for node, degree in G.degree() if degree > 3]
print(nodes_degree_gt_3)
```
